```r
x<-c(10,20,30,40,50)
y<-c(9.7,23.1,37.4,46.2,42.6)
```

```r
plot(x,y,col="blue", main="Population vs. Population 1 Year Later",xlab="Population P (in thousands)", ylab="Population f(P), 1 Year Later (in thousands)")
```

```r
relation3<-lm(y~x+I(x^2)+I(x^3))
print(relation3)
summary(relation3)
```

```r
x_seq <- seq(from=0,to=50, by=0.1)
new_relation3<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation3[i]<-predict(relation3, x_new)
}
plot(x,y,col="blue", main="Linear Function",lines(x_seq, new_relation3),xlab="x",ylab="y")
```

```r
confint(relation3, level=0.95)
```

```r
plot(fitted(relation3), residuals(relation3))
```
